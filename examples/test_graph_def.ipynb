```python
import tensorflow.compat.v1 as tf
import lucid_kietzmannlab.modelzoo.vision_models as models
tf.compat.v1.disable_eager_execution()
```

```python
model_checkpoint_dir="/Users/vkapoor/Downloads/models/AlexNet/training_seed_05"
training_seed = 5
model_ecoset = models.AlexNetv2(model_checkpoint_dir=model_checkpoint_dir, training_seed = training_seed)
t_input_ecoset = tf.compat.v1.placeholder(tf.float32, [None, *model_ecoset.image_shape])
scope = ""
tf.compat.v1.import_graph_def(
            model_ecoset.graph_def, {model_ecoset.input_name:t_input_ecoset}, name=scope
        )
graph = tf.compat.v1.get_default_graph()
output_node_names_ecoset = [
                    n.name
                    for n in graph.as_graph_def().node
                ]
layer_name_list = [layer_info["name"] for layer_info in model_ecoset.layers]
print(t_input_ecoset.shape)
print(output_node_names_ecoset)
print(layer_name_list)
common_elements = list(set(layer_name_list) & set(output_node_names_ecoset))
print(common_elements)
```

```python
for name in output_node_names_ecoset:
    if 'Conv2D' in name:
       print(name)
```

```python
for layer_name in layer_name_list:
   tensor_shape = graph.get_tensor_by_name(
                        f"{layer_name}:0"
                    ).shape
   print(tensor_shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color
from scipy.misc import face
from lucid_kietzmannlab.utils import generate_gabor_kernels, apply_gabor_kernels, extract_gabor_orientation, plot_gabor_kernels

```

```python
image = face()
image_gray = color.rgb2gray(image)
frequencies = [0.1, 0.2, 0.3, 0.4]
orientations = np.linspace(0, np.pi, 8, endpoint=False)  # 8 orientations from 0 to 180 degrees

# Generate and apply Gabor kernels
kernels = generate_gabor_kernels(frequencies, orientations)
responses = apply_gabor_kernels(image_gray, kernels)
plot_gabor_kernels(frequencies,kernels)
# Extract Gabor orientations
orientation_map = extract_gabor_orientation(responses, orientations)

# Display the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_gray, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Gabor Orientation Map")
plt.imshow(orientation_map, cmap='hsv')
plt.colorbar()
plt.show()

```
